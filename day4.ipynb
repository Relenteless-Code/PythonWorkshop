```python
#WAP to reversee a string

def reverse(str):
    start = 0
    end = len(str) - 1
    str1 = ''
    while start <= end:
      str1 += str[end]
      end = end - 1
    return str1
str1 = 'abcd'
print(reverse(str1))
```

```python
#Lambda function

even = lambda n: n%2 == 0

for i in range(1,20):
    if even(i):
        print(i)
```

```python
#WAP to find the maximum of three numbers using a lambda function
max = lambda a,b,c : a if a>b and a>c else b if b>c else c
print(max(1,4,3))
print(max(3,2,1))
print(max(1,3,2))
                                            
```

```python

def recursive_reverse(s):
    if len(s) == 0:
        return s
    else:
        return recursive_reverse(s[5:]) + s[0]

result = recursive_reverse("python")
print(result)
```
