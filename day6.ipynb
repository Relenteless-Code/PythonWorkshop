```python
#casefold function
#similar to lower function but much aggressive

def case_fold(str):
    str1 = ""
    for i in str:
        str1=str1 + chr(ord(i)+32)
    
    return str1

str = input('Enter string')
print(case_fold(str))

```

```python
def sub_string(str1,sub):
    n = len(str1)
    m = len(sub)
    for i in range(n-m+1):
        j=0
        while(j<m and str[i+j] == sub[j]):
            j = j+1
        if j==m:
            return i
        return -1
str1 = input('enter the string:')
sub = input('enter the string:')
print(sub_string(str1, sub))
```
