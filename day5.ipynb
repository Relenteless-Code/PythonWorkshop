```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
def create(initial_balance):
    balance = initial_balance
    def transaction(amount):
        nonlocal balance
        balance+=amount
        return balance
    return transaction

account = create(100)
print(account(50))
print(account(-30))
```

```python
def do_twice(func):
    def wrapper_do_twice(*args, **kwargs):
        func(*args, **kwargs)
        func(*args, **kwargs)
    return wrapper_do_twice

@do_twice
def message(name):
    print('hello', name)   # or print('hello ' + name)

name = input('enter name: ')
message(name)
```

GENERATORS IN PYTHON

1. A python generator is a kind of an iterable, like a python list or tuple.
2. It generates a sequence of values that we can Iterate on
3. Memory efficient: unlike lists, they don’t store all values in memory.

```python
def square(n):
    list1 = []
    for i in range(1, n+1):
        list1.append(i*i)
    return list1
num = int(input('Enter the number'))
print(square(num))
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True   

def generate_prime(start, end):
    for num in range(start, end + 1):
        if is_prime(num):
            yield num


prime_generator = generate_prime(5, 50)

for prime in prime_generator:
    print(prime, end=", ")
```

```python
#print all characters of the passed string in reverse order

def rev_str(my_str):
    lenght = len(my_str)
    for i in range(lenght -1, -1, -1):
        yield my_str[i]

for char in rev_str('hello'):
    print(char)
```

```python
def my_gen(x):
    while(x>0):
        if x%2==0:
            yield 'Even'
        else:
            yield 'Odd'
        x -= 1

for i in my_gen(5):
    print(i)
```
