```python
def arthimetic(n1,n2):
    add = n1+n2
    sub = n1-n2
    multiply = n1*n2
    div = n1/n2
    return add,sub,multiply,div

a,b,c,d = arthimetic(10,2)
print(a,b,c,d)
```

```python
#WAP to find factorial using recursion
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    else:
        # Recursive case
        return n * factorial(n - 1)

num = int(input("Enter a number: "))
print("Factorial of", num, "is:", factorial(num))
```

```python
# WAP to find exponential using recursion

def power(base, exp):
    # Base case
    if exp == 0:
        return 1
    else:
        # Recursive case
        return base * power(base, exp - 1)

# Example
b = int(input("Enter base: "))
e = int(input("Enter exponent: "))

print(f"{b}^{e} =", power(b, e))
```

```python
#user dfined function
def absolute(n):
    if n < 0:
        return - n
    else:
        return n
print(absolute(10))
print(absolute(-10))
```

```python
#find max element from a list
def largest():
    l = [1,23,34,53,2,9]
    max=0
    max=l[0]
    for i in range(0,len(l)):
        if max <l[i]:
            max=l[i]
    return max
maximum = largest()
print(maximum)
```

```python
#find the minimum element in an array

def smallest():
    l=[212,2,34,35,52,42]
    min = l[0]
    for i in range(0,len(l)):
        if min>l[i]:
            min = l[i]
    return min

minimum = smallest()
print(minimum)
```

```python
#second smallest element in an array

def small():
    l=[23,3,345,34,4,2,45]
    l.sort()
    return l[1]
min2 = small()
print(min2)
```

```python
#check whether a string is palindrome or not

str = input('enter string')
start = 0
end = len(str)-1
counter = 0
while start < end:
    if str[start] == str[end]:
        start = start +1
        end = end -1 
        continue
    else:
        counter = counter + 1
        break
if counter == 0:
    print('Palindrome')
else:
    print('Not Palindrome')
```
